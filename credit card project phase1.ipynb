Detecting Missing Values
df.isnull() â€“ Checks for nulls

df.isnull().sum() â€“ Total nulls per column

df.info() â€“ Overview of nulls and data types

âœ… Methods to Handle Missing Values
1. Removing
Drop rows: df.dropna()

Drop columns: df.dropna(axis=1)

âœ… Use if missing values are few and data loss is minimal.

2. Imputation (Filling)
Mean: df['col'].fillna(df['col'].mean())

Median: df['col'].fillna(df['col'].median())

Mode: df['col'].fillna(df['col'].mode()[0])

Forward Fill: df.fillna(method='ffill')

Backward Fill: df.fillna(method='bfill')

Custom Value: df.fillna(0) or 'Unknown'

Interpolate (numerical): df.interpolate()

```python
def handling_outliers(df): #handling outliers
    for col in df.select_dtypes(include='number').columns:
        Q1=df[col].quantile(0.25)
        Q3=df[col].quantile(0.75)
        IQR=Q3-Q1
        lower_bound=Q1-(1.5*IQR)
        upper_bound=Q3+(1.5*IQR)
        #Capping
        df[col]=df[col].apply(lambda x:lower_bound if x<lower_bound else upper_bound
                              if x>upper_bound else x)
    return df
    
        
```
